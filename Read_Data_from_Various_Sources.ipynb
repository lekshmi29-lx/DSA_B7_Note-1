**Reading Data from Various Sources**

Data can come from CSV, Excel, JSON, SQL Databases, APIs, Big Data sources.

Python libraries: *pandas.read_csv(), pandas.read_excel(), pd.read_json(), sqlalchemy,* etc.


**Array**

Arrays are homogeneous data structures (same type of elements).

In Python: *numpy.array()* is used for fast computation.

Useful for mathematical operations, linear algebra, statistics.

**DataFrame**

Two-dimensional labeled data structure (rows + columns).

In Pandas: *pd.DataFrame().*

**Series**

A one-dimensional labeled array in Pandas.

Holds data of any type (int, float, string, etc.).

Example: *pd.Series([10, 20, 30], index=['a','b','c'])*.

Useful for time-series and single column analysis.

**API**

When reading data from an API, the format of the data can vary. As seen from the output, the API returned a JSON object, which was then converted into a pandas DataFrame. Other APIs might return data in different formats like XML or CSV. You can use libraries like requests to fetch data from APIs and then use appropriate pandas functions like read_json, read_xml, or read_csv to load the data into a DataFrame depending on the format.

```python
import requests
url ="https://disease.sh/v3/covid-19/countries"

response = requests.get(url)
data = response.json()
#data
```

```python
covid_df = pd.DataFrame(data)
covid_df.head()
```

```python
url = "https://api.spacexdata.com/v4/launches"
resp = requests.get(url)
launches = resp.json()
#launches
```

```python
launch = pd.DataFrame(launches)
launch.head()
```

```python
df1_selected = launch[['name','date_utc','success','details','rocket','launch_library_id']]
df1_selected
```

```python
df1_selected= df1_selected.rename(columns={"name":"mission_name","rocket":"rocket_id"})
df1_selected
```

```python
url = "https://api.spacexdata.com/v4/rockets"

res = requests.get(url)
rockets = res.json()
rockets
```

```python
rocket = pd.DataFrame(rockets)
rocket.head()
```

```python
df2_selected = rocket[['id','name','country','company','cost_per_launch']]
df2_selected
```

```python
final_data = df1_selected.merge(df2_selected,left_on='rocket_id',right_on='id',how ='left')
final_data
```

```python
final_data.shape
```

```python
final_data = df1_selected.merge(df2_selected,left_on='rocket_id',right_on='id',how ='inner')
final_data.shape
```
