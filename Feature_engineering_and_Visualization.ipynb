# **Feature Engineering**

The art of transforming raw data into features that **make machine learning models work better**.


## Why Feature Engineering?

* Raw data is often **messy, incomplete, or unstructured**.
* ML algorithms require **clean, numerical, informative inputs**.
* Good features = higher model accuracy.

---

## **Types of Feature Engineering**

### **Feature Creation (New Features from Existing Data)**

* Example: In a hotel booking dataset →

  * Given: `CheckInDate`, `CheckOutDate`.
  * New Feature: `StayDuration = CheckOutDate - CheckInDate`.

* Example: E-commerce dataset →

  * Given: `Price`, `Discount`.
  * New Feature: `FinalPrice = Price - Discount`.

---

### **Feature Transformation (Change Scale / Distribution)**

* **Normalization / Scaling**: Put all features in same scale.

  * Example: `Income` (₹10,000 – ₹10,00,000) vs `Age` (20 – 70).
  * Use `MinMaxScaler` or `StandardScaler`.

* **Log Transformation**: Handle skewed distributions.

  * Example: House prices → use `log(price)` to reduce extreme variations.

---

### **Feature Encoding (Convert Categorical → Numerical)**

* **Label Encoding**: Assign numbers.

  * Example: City = {Delhi=0, Mumbai=1, Chennai=2}.
* **One-Hot Encoding**: Binary columns for each category.

  * Example:

    * Category: {Electronics, Clothing} →
      `Electronics=1, Clothing=0`

---

### **Feature Selection (Choose Best Features)**

* **Remove redundant or irrelevant features**.
* Example: If `TotalMarks` and `Percentage` exist → keep one.
* Use techniques like **Correlation, Chi-Square Test, PCA, Feature Importance**.

---

## **Example: Raw vs Engineered Dataset**

### **Raw Dataset (E-commerce Orders)**

| OrderID | Price | Discount | DeliveryDate | Category    |
| ------- | ----- | -------- | ------------ | ----------- |
| 101     | 500   | 50       | 2025-09-20   | Electronics |

### **After Feature Engineering**

| OrderID | FinalPrice | DaysToDeliver | Category_Electronics | Category_Clothing |
| ------- | ---------- | ------------- | -------------------- | ----------------- |
| 101     | 450        | 5             | 1                    | 0                 |

# **Data Visualization**

The process of **graphically representing data** to spot patterns, trends, and insights.
## **Why Visualization?**

* Helps in **EDA (Exploratory Data Analysis)**.
* Detects **outliers, correlations, distributions**.
* Easier to explain findings in reports & dashboards.

---

## **Visualization Techniques by Data Type**

### **For Numerical Data (Continuous)**

* **Histogram** → Show distribution of values.

  * Example: Income distribution of customers.
* **Boxplot** → Detect outliers.

  * Example: House prices with extreme high values.
* **Line Chart** → Trend over time.

  * Example: Monthly sales growth.

---

### **For Categorical Data**

* **Bar Chart** → Frequency count.

  * Example: Number of orders per city.
* **Pie Chart** → Proportions.

  * Example: Market share of product categories.

---

### **Numerical vs Categorical**

* **Boxplot / Violin Plot**

  * Example: Salary distribution across job titles.

---

### **Numerical vs Numerical**

* **Scatter Plot**

  * Example: Relationship between house size and price.
* **Heatmap**

  * Example: Correlation between features (age vs income vs spending score).

---

## **Example: House Price Dataset**

| HouseID | Size (sqft) | Price (₹ Lakhs) | City    |
| ------- | ----------- | --------------- | ------- |
| 1       | 1200        | 80              | Delhi   |
| 2       | 2000        | 150             | Mumbai  |
| 3       | 900         | 50              | Chennai |

### **Visualizations**

* **Histogram (Price)** → See if prices are skewed.
* **Scatter Plot (Size vs Price)** → Bigger houses → higher price?
* **Bar Chart (City vs Avg Price)** → Which city has higher prices?
* **Heatmap** → Correlation between Size & Price.

---

##  **Summary**

* **Feature Engineering** = Making raw data into useful model features.

  * Includes creation, transformation, encoding, selection.
* **Visualization** = Graphical exploration of data.

  * Helps in understanding distributions, relationships, and patterns.
